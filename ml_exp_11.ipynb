```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data       # shape (150, 4)
y = iris.target     # class labels (0,1,2)

# Standardize features (important for PCA)
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# --- PCA ---
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

print("Explained variance ratio by PCA components:", pca.explained_variance_ratio_)

# --- t-SNE ---
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_std)

# --- Visualization ---
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# PCA plot
scatter1 = axs[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
axs[0].set_title("PCA (2 components)")
axs[0].set_xlabel("PC1")
axs[0].set_ylabel("PC2")
legend1 = axs[0].legend(*scatter1.legend_elements(), title="Classes")

# t-SNE plot
scatter2 = axs[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
axs[1].set_title("t-SNE (2 components)")
axs[1].set_xlabel("t-SNE1")
axs[1].set_ylabel("t-SNE2")
legend2 = axs[1].legend(*scatter2.legend_elements(), title="Classes")

plt.show()
```
